```python
from sklearn.model_selection import train_test_split
features = finish.iloc[:, :-4]
labels = finish['labelWIN']


train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.01)
```

```python
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error

label_encoder = LabelEncoder()

train_labels_encoded = label_encoder.fit_transform(train_labels)
test_labels_encoded = label_encoder.transform(test_labels)

model = xgb.XGBClassifier(tree_method="gpu_hist", enable_categorical=True)
model.fit(train_features, train_labels_encoded) 

pred_labels_encoded = model.predict(test_features)
pred_labels = label_encoder.inverse_transform(pred_labels_encoded)

accuracy = accuracy_score(test_labels, pred_labels)
print('accuracy:',accuracy)
# mse = mean_absolute_error(test_labels,pred_labels)
# print(mse)
```

```python
import tensorflow as tf
from tensorflow.keras import layers

# Define the model
model2 = tf.keras.Sequential([
    layers.Dense(64, activation='tanh', input_shape=(71,)),
    layers.Dense(3)
])

optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
# Compile the model
model2.compile(optimizer=optimizer,
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])

# Train the model
model2.fit(train_features, train_labels_encoded, epochs=50)

# Evaluate the model
test_loss, test_acc = model2.evaluate(test_features, test_labels_encoded)
print('Test accuracy:', test_acc)
```
